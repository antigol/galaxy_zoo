```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot(file, label="", color="red"):
    data = np.loadtxt(file)
    ite = data[:,0]
    test = data[:,1]
    train = data[:,2]
    plt.plot(ite, test, label=label, color=color)
    plt.plot(ite, train, '.', color=color)
```

```python
#plot('runs/bl_b1/metrics.txt', 'baseline', color="red")
#plot('runs/bl_b2/metrics.txt', 'baseline', color="red")
plot('runs/inv_b1/metrics.txt', 'inv', color="green")
#plot('runs/inv_b2/metrics.txt', 'inv', color="green")
plot('runs/new_baseline/metrics.txt', 'new baseline', color="blue")
#plot('runs/new_invariant/metrics.txt', 'new invariant', color="orange")
plot('runs/new_invariant_a1/metrics.txt', 'new invariant a', color="black")
plot('runs/new_invariant_b1/metrics.txt', 'new invariant b', color="red")
plot('runs/new_invariant_c1/metrics.txt', 'new invariant c', color="orange")
plot('runs/new_invariant_d1/metrics.txt', 'new invariant d', color="brown")
plot('runs/new_invariant_e1/metrics.txt', 'new invariant e', color="cyan")
plt.legend()
plt.ylim([0.0749, 0.15])
```
